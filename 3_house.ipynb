```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/Users/vidhanmanihar/Desktop/untitled folder/DSML/Practical/three/House_Price.csv')
```

```python
print("Dataset Loaded Successfully!")
print(df.head())
```

```python
numeric_cols = df.select_dtypes(include=[np.number]).columns
print("Numeric Columns:", list(numeric_cols))
```

```python
print("------ Standard Deviation ------")
print(df[numeric_cols].std(numeric_only=True))
```

```python
print("------ Variance ------")
print(df[numeric_cols].var(numeric_only=True))
```

```python
print("------ Percentiles (25th, 50th, 75th) ------")
percentiles = df[numeric_cols].quantile([0.25, 0.50, 0.75])
print(percentiles)
print("\n---------------------------------------\n")
```

```python
print("Generating histograms...")

for col in numeric_cols:
    plt.figure(figsize=(6,4))
    plt.hist(df[col].dropna(), bins=30)
    plt.title(f"Histogram of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.grid(False)
    plt.show()
```
