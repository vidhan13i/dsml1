```python
import pandas as pd
import math
```

```python
df = pd.read_csv('/Users/vidhanmanihar/Desktop/untitled folder/DSML/Practical/four/Lipstick.csv')
```

```python
df = df.drop(columns=['Id'])    
```

```python
# Target column (last column = Buys)
target = df.columns[-1]
```

```python

# ------- ENTROPY FUNCTION -------
def entropy(col):
    values = col.value_counts()
    total = len(col)
    ent = 0
    for count in values:
        p = count / total
        ent -= p * math.log2(p)
    return ent
```

```python
# ------- INFORMATION GAIN -------
def information_gain(df, attribute, target):
    total_entropy = entropy(df[target])
    values = df[attribute].unique()
    weighted_entropy = 0

    for val in values:
        subset = df[df[attribute] == val]
        w = len(subset) / len(df)
        weighted_entropy += w * entropy(subset[target])

    ig = total_entropy - weighted_entropy
    return ig
```

```python
# Compute IG for each attribute except target
attributes = [col for col in df.columns if col != target]

print("Information Gain for each attribute:\n")

ig_values = {}

for attr in attributes:
    ig = information_gain(df, attr, target)
    ig_values[attr] = ig
    print(f"{attr} → IG = {ig:.4f}")
```

```python
# Find best attribute (root)
root = max(ig_values, key=ig_values.get)
print("\n==============================")
print(f"ROOT NODE of Decision Tree = {root}")
print("==============================")
```
