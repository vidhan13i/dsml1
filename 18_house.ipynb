```python
import pandas as pd

# Load dataset
df = pd.read_csv("House_Price.csv")

# Display first few rows
df.head()
```

```python
# Categorical variables
categorical_vars = df.select_dtypes(include=['object']).columns
print("Categorical Variables:\n", categorical_vars)

# Quantitative variables
quantitative_vars = df.select_dtypes(include=['int64', 'float64']).columns
print("\nQuantitative Variables:\n", quantitative_vars)
```

```python
quant_summary = df[quantitative_vars].describe().T
quant_summary
```

```python
cat_summary = df[categorical_vars].describe().T
cat_summary
```

```python
# Create age groups from numeric Age column
df["AgeGroup"] = pd.cut(
    df["Age"],
    bins=[0, 18, 30, 45, 60, 100],
    labels=["0-18", "19-30", "31-45", "46-60", "60+"]
)

# Grouped summary
grouped_income = df.groupby("AgeGroup")["Income"].agg(
    ["mean", "median", "min", "max", "std", "count"]
)

print(grouped_income)
```
