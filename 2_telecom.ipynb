```python
import pandas as pd
```

```python
df= pd.read_csv('/Users/vidhanmanihar/Desktop/untitled folder/DSML/Practical/two/Telecom_Churn.csv')
```

```python
print("Minimum values")
print(df.min(numeric_only=True))
```

```python
print("------ Maximum Values ------")
print(df.max(numeric_only=True))
```

```python
print("------ Mean Values ------")
print(df.mean(numeric_only=True))
```

```python
print("------ Range Values ------")

# Select only pure numeric columns (exclude booleans)
num_df = df.select_dtypes(include=['int64', 'float64'])

range_values = num_df.max() - num_df.min()
print(range_values)
```

```python
print("------ Standard Deviation ------")
print(df.std(numeric_only=True))
```

```python
print("------ Variance ------")
print(df.var(numeric_only=True))
```

```python
print("------ Percentiles (25th, 50th, 75th) - numeric cols only ------")
num_df = df.select_dtypes(include=['int64', 'float64'])   # exclude booleans
print(num_df.quantile([0.25, 0.50, 0.75]))
```

```python
print("------ Full Describe Summary ------")
print(df.describe())
```
