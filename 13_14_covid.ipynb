```python
import pandas as pd

# Load dataset
df = pd.read_csv("Covid Vaccine Statewise.csv")
```

```python
# (a) Describe the dataset
print("=== Dataset Description ===")
print(df.describe(include='all'))
```

```python

# (b) Number of persons state-wise vaccinated for FIRST dose in India
print("\n=== State-wise 1st Dose Vaccination Count ===")
first_dose = df.groupby("State")["First Dose Administered"].sum()
print(first_dose)
```

```python

# (c) Number of persons state-wise vaccinated for SECOND dose in India
print("\n=== State-wise 2nd Dose Vaccination Count ===")
second_dose = df.groupby("State")["Second Dose Administered"].sum()
print(second_dose)
```

```python
# (B) Number of males vaccinated (sum of column)
print("\n=== Total Males Vaccinated ===")
males = df["Male(Individuals Vaccinated)"].sum()
print(males)
```

```python
# (C) Number of females vaccinated (sum of column)
print("\n=== Total Females Vaccinated ===")
females = df["Female(Individuals Vaccinated)"].sum()
print(females)
```
