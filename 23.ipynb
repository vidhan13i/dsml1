```python
import pandas as pd
import numpy as np
from math import log2
```

```python
data = {
    "Age": ["Young", "Young", "Middle", "Old", "Old", "Old", "Middle", 
            "Young", "Young", "Old", "Young", "Middle", "Middle", "Old"],
    
    "Income": ["High", "High", "High", "Medium", "Low", "Low", "Low",
               "Medium", "Low", "Medium", "Medium", "Medium", "High", "Medium"],
    
    "Married": ["No", "No", "No", "No", "Yes", "Yes", "Yes",
                "No", "Yes", "Yes", "Yes", "No", "Yes", "No"],
    
    "Health": ["Fair", "Good", "Fair", "Fair", "Fair", "Good", "Good",
               "Fair", "Fair", "Fair", "Good", "Good", "Fair", "Good"],
    
    "Class": ["No", "No", "Yes", "Yes", "Yes", "No", "Yes",
              "No", "Yes", "Yes", "Yes", "Yes", "Yes", "No"]
}

df = pd.DataFrame(data)
df
```

```python
freq_table = df["Age"].value_counts()
print("Frequency Table for Age:\n")
print(freq_table)
```

```python
def entropy(col):
    values, counts = np.unique(col, return_counts=True)
    probs = counts / len(col)
    return -np.sum(probs * np.log2(probs))
```

```python
total_entropy = entropy(df["Class"])
print("Entropy of full dataset:", total_entropy)
```

```python
IG_age = total_entropy

# Weighted entropy after split
for age_group, subset in df.groupby("Age"):
    weight = len(subset) / len(df)
    ent = entropy(subset["Class"])
    IG_age -= weight * ent
    print(f"Entropy(Age = {age_group}): {ent:.4f}, weight = {weight:.4f}")

print("\nInformation Gain for splitting on Age =", IG_age)
```
